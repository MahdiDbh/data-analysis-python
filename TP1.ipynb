# Exercice 2 - Analyse du dataset Employment

```python
# Chargement du fichier employment
df_emp = pd.read_csv("employment.csv")
print("Fichier chargé | Taille :", df_emp.shape)
```

```python
# Afficher les 4 premières lignes
df_emp.head(4)
```

```python
# Vérifier les valeurs manquantes
print("-- Valeurs Manquantes --")
pd.DataFrame(df_emp.isnull().sum(), columns=['Nombre de valeurs manquantes'])
```

```python
# Vérifier les doublons
print("Nombre de doublons :", df_emp.duplicated().sum())

# Supprimer les doublons si il y en a
df_emp.drop_duplicates(inplace=True)
print("Doublons supprimés ")
```

```python
# Remplir les valeurs manquantes avec la moyenne
df_emp.fillna(df_emp.mean(numeric_only=True), inplace=True)
print("Valeurs manquantes traitées ")
print("Vérification :", df_emp.isnull().sum().sum(), "valeurs manquantes restantes")
```

```python
# Détection des outliers avec la méthode IQR
df_emp_num = df_emp.select_dtypes(include=[np.number])

Q1 = df_emp_num.quantile(0.25)
Q3 = df_emp_num.quantile(0.75)
IQR = Q3 - Q1

outliers = ((df_emp_num < (Q1 - 1.5 * IQR)) | (df_emp_num > (Q3 + 1.5 * IQR)))

# Créer un beau tableau récapitulatif
tableau_outliers = pd.DataFrame({
    'Q1'              : round(Q1, 2),
    'Q3'              : round(Q3, 2),
    'IQR'             : round(IQR, 2),
    'Limite Basse'    : round(Q1 - 1.5 * IQR, 2),
    'Limite Haute'    : round(Q3 + 1.5 * IQR, 2),
    'Nb Outliers'     : outliers.sum(),
})

print("-- DONNÉES ABERRANTES (OUTLIERS) --\n")
tableau_outliers
```

```python
# Plot des 3 secteurs d'emploi
plt.figure(figsize=(14, 6))

plt.plot(df_emp['Construction'], label='Construction', color='blue')
plt.plot(df_emp['Finance'], label='Finance', color='green')
plt.plot(df_emp['Education and Health'] , label='Education and Health', color='red')

plt.title("Taux d'emploi par secteur")
plt.xlabel("Index")
plt.ylabel("Taux d'emploi")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Analyse statistique complète sur Finance
finance = df_emp['Finance']

print("-- Analyse Statistique - FINANCE --\n")




print("Moyenne      :", round(finance.mean(), 2))
print("Médiane      :", round(finance.median(), 2))
print("Écart-type   :", round(finance.std(), 2))
print("Variance     :", round(finance.var(), 2))
print("Min          :", round(finance.min(), 2))
print("Max          :", round(finance.max(), 2))
print("Skewness     :", round(finance.skew(), 2))
print("Kurtosis     :", round(finance.kurtosis(), 2))

print("\n--- Interprétation ---")
if finance.skew() > 0:
    print(" Distribution asymétrique à DROITE (quelques très hautes valeurs)")
elif finance.skew() < 0:
    print(" Distribution asymétrique à GAUCHE (quelques très basses valeurs)")
else:
    print(" Distribution symétrique")
```

```python
# Corrélation entre Finance et toutes les autres variables
print("=== CORRÉLATION FINANCE VS AUTRES VARIABLES ===\n")

df_emp_num = df_emp.select_dtypes(include=[np.number])
correlation_finance = df_emp_num.corr()['Finance'].sort_values(ascending=False)
print(correlation_finance)

# Visualisation
plt.figure(figsize=(10, 6))
correlation_finance.drop('Finance').plot(kind='bar', color='steelblue', edgecolor='black')
plt.title("Corrélation entre Finance et les autres variables")
plt.xlabel("Variables")
plt.ylabel("Corrélation")
plt.axhline(y=0, color='red', linestyle='--')
plt.tight_layout()
plt.show()
```
